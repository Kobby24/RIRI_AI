```python
import torchaudio
from transformers import AutoProcessor, SeamlessM4Tv2Model
```

```python
# 1. Load the Processor and Model

# We use the 'large' model here, but you can swap to 'medium' to save RAM
model_id = "facebook/seamless-m4t-v2-large"
print(f"Loading {model_id}...")
processor = AutoProcessor.from_pretrained(model_id)
model = SeamlessM4Tv2Model.from_pretrained(model_id)
```

```python
# 2. Load your Twi audio file
# Upload your file to Colab's file explorer and replace the filename below
audio_file_path = "twi.wav"
audio_sample, orig_freq = torchaudio.load(audio_file_path)
```

```python
audio_sample = torchaudio.functional.resample(
    audio_sample,
    orig_freq=orig_freq,
    new_freq=16000
)
```

```python
audio_inputs = processor(audio=audio_sample, return_tensors="pt", sampling_rate=16000)
```

```python
# tgt_lang="eng" specifies English. generate_speech=False tells it to only output text.
print("Translating...")
output_tokens = model.generate(**audio_inputs, tgt_lang="eng", generate_speech=False)
```

```python
translated_text = processor.decode(output_tokens[0].tolist()[0], skip_special_tokens=True)

print("\n--- Result ---")
print("English Translation:", translated_text)
```
